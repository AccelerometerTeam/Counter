```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv("counterLog18112020.csv")
test = test.set_index(pd.to_datetime(test['Time']))
```

```python
fig, ax = plt.subplots(figsize=(16,4))
(test['2020-11-18 23:00':'2020-11-19 07:20']['Freq']-(test['2020-11-18 23:00':'2020-11-19 07:20']['Freq']).mean()).plot()
ax.set_ylabel('Offset Frequency from 6.3 GHz [Hz]')
```
